### Exploratory Data Analysis - Retail

**In this task, we perform exploratory data analysis on the sample superstore dataset**

```python
#import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data

store_df = pd.read_csv('SampleSuperstore.csv')
store_df.head()
```

```python
store_df.tail()
```

**Data Inspection (Cleaning)**

```python
#datatypes and null values

store_df.info()
```

```python
#find null values

store_df.isnull().sum()
```

```python
#description of data

store_df.describe()
```

```python
#check for duplicates

store_df.duplicated().sum()
```

```python
#remove duplicates

store_df.drop_duplicates(inplace = True)
store_df.duplicated().sum()
```

```python
#number of instances

store_df.shape
```

```python
# drop the unnecessary columns like postal code which doesn't affect the sales and profit

store_df.drop(columns="Postal Code", axis=1, inplace=True)
store_df.describe()
```

```python
store_df.head()
```

```python
#find correlation

corr1 = store_df.corr()
corr1
```

```python
#heatmap to check to correlation

sns.heatmap(corr1, annot = True)
```

**Exploratory Data Analysis**

```python
#view numerical data as mean value grouped by State column

store_df.state = store_df.groupby("State")[["Quantity", "Discount", "Sales", "Profit"]].mean()
store_df.state
```

```python
#view numerical data as median value grouped by State column

store_df.state = store_df.groupby("State")[["Quantity", "Discount", "Sales", "Profit"]].median()
store_df.state
```

```python
#view numerical data as min value grouped by State column

store_df.state = store_df.groupby("State")[["Quantity", "Discount", "Sales", "Profit"]].min()
store_df.state
```

```python
#view numerical data as max value grouped by State column

store_df.state = store_df.groupby("State")[["Quantity", "Discount", "Sales", "Profit"]].max()
store_df.state
```

```python
#find sales profit by state

sales_profit = store_df.groupby("State")["Profit"].sum()
sales_profit
```

```python
#plot sales profit

plt.figure(figsize = (20,10))
sns.barplot(x=sales_profit.index, y=sales_profit.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

```python
#pair plot by region

sns.pairplot(store_df, hue="Region", diag_kind="hist")
```

```python
#pair plot by segment

sns.pairplot(store_df, hue="Segment", diag_kind="hist")
```

```python
#pair plot by category

sns.pairplot(store_df, hue="Category", diag_kind="hist")
```

```python
#pair plot by sub-category

sns.pairplot(store_df, hue="Sub-Category", diag_kind="hist")
```

```python
sns.countplot("Region", data = store_df)
```

```python
sns.countplot("Segment", data = store_df)
```

```python
sns.countplot("Category", data = store_df)
```

```python
plt.xticks(rotation = 90)
sns.countplot("Sub-Category", data = store_df)
```

**Sales by States**

```python
sales = store_df.groupby("State")["Sales"].sum().sort_values(ascending=False)
sales.plot.bar(figsize=(20,10))
```

**Sales Profit per Category**

```python
category = store_df.groupby("Sub-Category")["Profit", "Sales"].sum()
category.plot.bar(figsize=(20,10))
```

**Profit By States**

```python
store_df.groupby("State").Profit.agg(["sum", "mean", "min", "max", "median", "count", "std", "var"])
```

```python
# Category vs. Region

pd.crosstab(store_df["Region"], store_df["Category"]).plot(kind="bar", figsize=(20,10))
plt.xticks(rotation=45)
plt.plot()
```

```python
# Sub-Category vs. Region

plt.figure(figsize = (20,10))
sns.countplot(x = "Sub-Category", hue = "Region", data = store_df)
plt.xticks(rotation = 90)
plt.plot()
```

**Conclusion:**

1. Many sales need improvements in terms of category and subcategory, they should be sold at a reasonable price.

2. Profits and sales are highly correlated.

3. There is no correlation between quantity and profit.

4. Segment requires more data as no correlation could be found.
